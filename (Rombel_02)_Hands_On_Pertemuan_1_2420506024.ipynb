### 3. Praktik PySpark Lanjutan
- **Latihan 1**: Memanipulasi Data dengan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
````

- **Tugas 2**: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

````python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana, dg tambahan kolom pekerjaan, hobi, & gender
data = [
    ("Ali", 34, "Dokter", "Olahraga", "Laki-laki"),
    ("Budi", 23, "Guru", "Traveling","Laki-laki"),
    ("Citra", 29, "Karyawan Swasta", "Memasak", "Perempuan"),
    ("Dina", 45, "Karyawan BUMN", "Shoping", "Perempuan")
    ]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
print("\033[1m                   DataFrame\033[0m")
df.show()

# Filtering data berdasarkan kolom Pekerjaan yg mengandung kata "Karyawan"
from pyspark.sql.functions import col
print("\033[1m                Filtering Data\033[0m")
df_filtered = df.filter(col("Pekerjaan").contains("Karyawan"))
df_filtered.show()

# Menghitung rata-rata usia
print("\033[1mRata-rata Usia\033[0m")
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data "Nama" secara descending
print("\033[1m        Mengurutkan Data Nama (Descending)\033[0m")
df_sorted = df.orderBy("Nama", ascending=False)
df_sorted.show()

# Mengurutkan data "Usia" secara ascending
print("\033[1m        Mengurutkan Data Usia (Ascending)\033[0m")
df_sorted = df.orderBy("Usia", ascending=True)
df_sorted.show()
````

### 5. Praktik Pandas Lanjutan

- **Tugas 5**: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

````python
# Data Spark
data_spark = [
    ("Ali", 34, "Dokter", "Olahraga", "Laki-laki"),
    ("Budi", 23, "Guru", "Traveling","Laki-laki"),
    ("Citra", 29, "Karyawan Swasta", "Memasak", "Perempuan"),
    ("Dina", 45, "Karyawan BUMN", "Shoping", "Perempuan")
    ]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data_spark, columns)

# Data Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Usia": [34, 23, 29, 45],
    "Domisili" : ["Surabaya", "Jambi", "Bandung", "Sleman"]
    }
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

# Menampilkan DataFrame hasil konversi
df_pandas_from_spark

# Menggabungkan DataFrame
df_joined = pd.merge(df_pandas, df_pandas_from_spark, on="Nama")

# Menghapus duplikasi kolom usia (Usia_y)
# dan merapikan nama kolom (Usia_x menjadi Usia)
df_joined = df_joined.drop(columns=['Usia_y'])
df_joined.rename(columns={'Usia_x': 'Usia'}, inplace=True)
print(df_joined)
print()

# Menghitung rata-rata usia
avg_usia = df_joined['Usia'].mean()
print("\033[1mRata-rata usia:\033[0m", avg_usia)
print()
````

### 6. Menggabungkan PySpark dan Pandas

- **Tugas 6**: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

````python
# Data Spark
data_spark = [
        ("Ali", 34, "Dokter", "Olahraga", "Laki-laki"),
    ("Budi", 23, "Guru", "Traveling","Laki-laki"),
    ("Citra", 29, "Karyawan Swasta", "Memasak", "Perempuan"),
    ("Dina", 45, "Karyawan BUMN", "Shoping", "Perempuan")
    ]
columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df = spark.createDataFrame(data_spark, columns)

# Data Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Usia": [34, 23, 29, 45],
    "Domisili" : ["Surabaya", "Jambi", "Bandung", "Sleman"]
    }
df_pandas = pd.DataFrame(data_pandas)

# Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

# Menampilkan DataFrame hasil konversi
df_pandas_from_spark

# Menggabungkan DataFrame
df_joined = pd.merge(df_pandas, df_pandas_from_spark, on="Nama")

# Menghapus duplikasi kolom usia (Usia_y)
# dan merapikan nama kolom (Usia_x menjadi Usia)
df_joined = df_joined.drop(columns=['Usia_y'])
df_joined.rename(columns={'Usia_x': 'Usia'}, inplace=True)
print(df_joined)
print()

# Nilai maksimum usia
max_usia = df_joined['Usia'].max()
print("\033[0mNilai maksimum usia:\033[0m", max_usia)
````
