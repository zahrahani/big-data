```python
# Exercise 1: Identifying and Handling Missing Data

import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}

df = pd.DataFrame(data)

# Filling missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)

# Dropping rows
df.dropna(subset=['Name'], inplace=True)

# Show data
print('\033[1mAfter cleaning:\033[0m\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data

import pandas as pd

# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}

df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()

# Show standardized data
print('\033[1mStandardized Data:\033[0m')
display(df)
```

```python
# Load dataset Titanic-Dataset.csv

import pandas as pd

df = pd.read_csv("/home/hani/Downloads/Titanic-Dataset.csv")

print('Dataset loaded successfully!')
```

```python
# Imputation: fill missing values in the 'Age' column with the mean value

df['Age'] = df['Age'].fillna(df['Age'].mean())

print('\033[1mJumlah Missing Value di Setiap Kolom (Setelah Imputasi):\033[0m')
df.isnull().sum()
```

```python
# Display a few rows of data after imputation

print('\033[1mBeberapa Baris Data Setelah Imputasi:\033[0m')
df[['Age', 'Fare']].head(10)
```

```python
# Normalize numerical data in the 'Age' and 'Fare' columns to a 0â€“1 range

df['Age_norm'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Fare_norm'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())

print('\033[1mData Setelah Normalisasi (Age & Fare):\033[0m')
display(df[['Age', 'Age_norm', 'Fare', 'Fare_norm']].head(10))
```

```python
# Standardize categorical columns

# Replace NaN values in the 'Cabin' column with 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)

# Standardize text format
df['Sex'] = df['Sex'].str.lower()
df['Embarked'] = df['Embarked'].fillna('U').str.upper()
df['Cabin'] = df['Cabin'].str.upper()

print('\033[1mHasil Standardisasi Kolom Kategorikal:\033[0m')
print(df[['Sex', 'Embarked', 'Cabin']].head(10))
```

```python
# Remove duplicate rows from the dataset

before = df.shape[0]  # number of rows before
df.drop_duplicates(inplace=True)
after = df.shape[0]   # number of rows after

print('\033[1mHasil Penghapusan Data Duplikat\033[0m')
print(f'Jumlah baris sebelum: {before}')
print(f'Jumlah baris sesudah : {after}')
print(f'Total duplikat yang dihapus: {before - after}')
```

```python
# Visualization: create a grouped bar chart comparing survival counts by passenger class

import matplotlib.pyplot as plt
import numpy as np

# Group the data by Pclass and Survived, then reshape it for plotting
group_data = df.groupby(['Pclass', 'Survived']).size().unstack()

plt.figure(figsize=(9, 6))

# Set positions and bar width
x = np.arange(len(group_data.index))
width = 0.35


# Create bars for 'Not Survived' and 'Survived'
bar1 = plt.bar(x - width/2, group_data[0], width, label='Not Survived', color='lightcoral')
bar2 = plt.bar(x + width/2, group_data[1], width, label='Survived', color='lightgray')

# Add value labels for 'Not Survived'
for b in bar1:
    plt.text(
        b.get_x() + b.get_width()/2,       
        b.get_height() + 2,                
        str(int(b.get_height())),          
        ha='center', va='bottom', fontsize=10
    )

# Add value labels for 'Survived'
for b in bar2:
    plt.text(
        b.get_x() + b.get_width()/2,
        b.get_height() + 2,
        str(int(b.get_height())),
        ha='center', va='bottom', fontsize=10
    )

# Add axis labels, title, and legend
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Survival Comparison Across Passenger Classes")
plt.xticks(x, ['1st Class', '2nd Class', '3rd Class'])
plt.legend()
plt.show()
```

```python
# Dashboard: multiple visualizaton summarizing the Titanic Dataset

import matplotlib.pyplot as plt
import seaborn as sns

# Set default visualization style
sns.set_style('whitegrid')

# Create a 2x2 grid layout for four plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plotting
# Plot 1: Histogram - Age distribution of passengers
sns.histplot(df['Age'], 
             bins=20, 
             color='steelblue', 
             ax=axes[0, 0])
axes[0, 0].set_title('Histogram: Age Distribution of Passengers')

# Plot 2: Pie chart - Distribution of embarkation points
embark_counts = df['Embarked'].value_counts()
axes[0, 1].pie(embark_counts, labels=embark_counts.index, 
               autopct='%1.1f%%', startangle=90, 
               colors=['#3d6f95', '#6ca0c8', '#9bbcd9', '#cfe0f1'])
axes[0, 1].set_title('Pie Chart: Embarkation Point Distribution')

# Plot 3: Horizontal grouped bar chart - Survival passengers by age group
# Categorize passengers into age groups
def age_group(age):
    if age < 13:
        return 'Child'
    elif age < 20:
        return 'Teen'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
# Apply function age_group to dataset
df['AgeGroup'] = df['Age'].apply(age_group) 
# Count survived vs not-survived for each age group
age_survival = df.groupby(['AgeGroup', 'Survived']).size().unstack(fill_value=0)
# Reorder the age group to keep them in a logical order
age_survival = age_survival.loc[['Child', 'Teen', 'Adult', 'Senior']]  
# Plot the horizontal grouped bar chart
age_survival.plot(kind='barh',
                  ax=axes[1, 0],
                  color=['#d4dce3', 'steelblue'])

axes[1, 0].set_title('Grouped Bar: Survival by Age Group')
axes[1, 0].set_xlabel('Passenger Count')
axes[1, 0].set_ylabel('Age Group')
axes[1, 0].legend(['Not Survived', 'Survived'], title='Status')
# Add number labels on each bar
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, padding=3)

# Plot 4: Survivors passengers by gender and embarkation
# Filter only survived passengers and count by Embarked and Sex
survival_by_group = (df[df['Survived'] == 1]
                     .groupby(['Embarked', 'Sex'])
                     .size()
                     .reset_index(name='SurvivorCount'))
# Plot the bar chart
sns.barplot(data=survival_by_group,
            x='Embarked', y='SurvivorCount',
            hue='Sex',
            palette=['lightcoral', 'steelblue'],
            ax=axes[1, 1])

axes[1, 1].set_title('Bar Chart: Survivors by Gender and Embarkation')
axes[1, 1].set_xlabel('Embarkation Point')
axes[1, 1].set_ylabel('Survivor Count')
# Add number labels on each other
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, padding=3)

# Final layout settings
plt.suptitle("Titanic Dataset Dashboard", fontsize=16)
plt.tight_layout()
plt.show()
```
