```python
# Koneksi MongoDB Atlas dengan Colab

from pymongo import MongoClient
import pandas as pd

url = "mongodb+srv://zahranibras:Makan1hari3x@learn-bigdata.6emxxsy.mongodb.net/"
client = MongoClient(url)

db = client.sample_airbnb
collection = db.listingsAndReviews

df = pd.DataFrame(list(collection.find()))
```

```python
# Tampilkan daftar kolom yang ada di dalam sample_airbnb

print(df.columns)
```

```python
# Filter dataset sample_airbnb
# Kriteria → room_type = "Entire home/apt"

df_room_type = df[df["room_type"] == "Entire home/apt"]

df_selected_columns = df_room_type[
    ["_id", "name", "room_type", "accommodates", "price"]
    ].rename(columns = {
        "_id" : "ID",
        "name": "Name",
        "room_type": "Room Type",
        "accommodates": "Accommodates",
        "price": "Price"
        }).reset_index(drop=True)

print("\033[1mFilter by room_type = 'Entire home/apt'\033[0m\n")
df_selected_columns.head(10)
```

```python
# Filter dataset sample_airbnb
# Kriteria → property_type = "House"

df_property_type = df[df["property_type"] == "House"]

df_selected_columns = df_property_type[
    ["_id", "name", "property_type", "amenities", "price"]
    ].rename(columns = {
        "_id" : "ID",
        "name": "Name",
        "property_type": "Property Type",
        "amenities": "Amenities",
        "price": "Price"
    }).reset_index(drop=True)

print("\033[1mFilter by property_type = 'House'\033[0m\n")
df_selected_columns.head(10)
```

```python
# Filter dataset sample_airbnb
# Kriteria → bed_type = "Real Bed"

df_bed_type = df[df["bed_type"] == "Real Bed"]

df_selected_columns = df_bed_type[
    ["_id", "name", "bed_type", "bedrooms", "accommodates", "price"]
    ].rename(columns = {
        "_id" : "ID",
        "name": "Name",
        "bed_type": "Bed Type",
        "bedrooms": "Bedrooms",
        "accommodates": "Accommodates",
        "price": "Price"
    }).reset_index(drop=True)

print("\033[1mFilter by bed_type = 'Real bed'\033[0m\n")
df_selected_columns.head(10)
```

```python
# Filter dataset sample_airbnb
# Kriteria → cancellation_policy= "flexible"

df_cancellation_policy = df[df["cancellation_policy"] == "flexible"]

df_selected_columns = df_cancellation_policy[
    ["_id", "name", "cancellation_policy", "property_type", "price"]
    ].rename(columns = {
        "_id" : "ID",
        "name": "Name",
        "cancellation_policy": "Cancellation Policy",
        "property_type": "Property Type",
        "price": "Price"
    }).reset_index(drop=True)

print("\033[1mFilter by cancellation_policy = 'flexible'\033[0m\n")
df_selected_columns.head(10)
```

```python
# Filter dataset sample_airbnb
# Kriteria → minimum pemesanan 2 malam (minimum_nights >= 2)

df["minimum_nights"] = df["minimum_nights"].astype(int)
df_minimum_nights = df[df["minimum_nights"] >= 2]

df_selected_columns = df_minimum_nights[
    ["_id", "name", "minimum_nights", "room_type", "price"]
    ].rename(columns = {
        "_id" : "ID",
        "name": "Name",
        "minimum_nights": "Minimum Nights",
        "room_type": "Room Type",
        "price": "Price"
    }).reset_index(drop=True)

print("\033[1mFilter by minimum booking 2 nights\033[0m\n")
df_selected_columns.head(10)
```
