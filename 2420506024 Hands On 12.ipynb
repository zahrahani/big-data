```python
# Exercise 1: Identifying and Handling Missing Data

import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}

df = pd.DataFrame(data)

# Filling missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)

# Dropping rows
df.dropna(subset=['Name'], inplace=True)

# Show data
print('\033[1mAfter cleaning:\033[0m\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data

import pandas as pd

# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}

df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()

# Show standardized data
print('\033[1mStandardized Data:\033[0m')
display(df)
```

```python
# Imputation: fill missing values in the 'Age' column with the mean value

df['Age'] = df['Age'].fillna(df['Age'].mean())

print('\033[1mJumlah Missing Value di Setiap Kolom (Setelah Imputasi):\033[0m')
df.isnull().sum()
```

```python
# Display a few roews of data after imputation

print('\033[1mBeberapa Baris Data Setelah Imputasi:\033[0m')
df[['Age', 'Fare']].head(10)
```

```python
# Normalize numerical data in the 'Age' and 'Fare' columns to a 01 range

df['Age_norm'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Fare_norm'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())

print('\033[1mData Setelah Normalisasi (Age & Fare):\033[0m')
display(df[['Age', 'Age_norm', 'Fare', 'Fare_norm']].head(10))
```

```python
# Standardize categorical columns

# Replace NaN values in the 'Cabin' column with 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)

# Standardize text format
df['Sex'] = df['Sex'].str.lower()
df['Embarked'] = df['Embarked'].fillna('U').str.upper()
df['Cabin'] = df['Cabin'].str.upper()

print('\033[1mHasil Standardisasi Kolom Kategorikal:\033[0m')
print(df[['Sex', 'Embarked', 'Cabin']].head(10))
```

```python
# Remove duplicate rows from the dataset

before = df.shape[0]  # number of rows before
df.drop_duplicates(inplace=True)
after = df.shape[0]   # number of rows after

print('\033[1mHasil Penghapusan Data Duplikat\033[0m')
print(f'Jumlah baris sebelum: {before}')
print(f'Jumlah baris sesudah : {after}')
print(f'Total duplikat yang dihapus: {before - after}')
```
